#### Облако слов (wordcloud).

```python
def create_word_cloud(df, text_col, title):
    """
    Создания облака слов
    
    params:
    df [pandas.DataFrame]: - датафрейм.
    
    text [str]: - столбец с текстом статьи.
    
    title [str]: - дополнительное уточнение в название графика.
    """
    text = " ".join(texts for texts in df[text_col])
    wordcloud = WordCloud(width=800, height=400, max_font_size=50, max_words=100, background_color="white").generate(text)
    
    plt.figure(figsize=(14, 8), dpi=80)
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Облако слов - {title}", fontsize=24)
    plt.show()

"""Стемминг"""
porter = PorterStemmer()
def tokenizer_porter(text):
    return [porter.stem(word) for word in text.split()]

"""Стоп-слова."""
def remove_stopwords_and_stem(text):
    word_array = [word for word in tokenizer_porter(text) if word not in stop]
    return ' '.join(word_array)

nltk.download('stopwords') # загрузка стоп-слов.
stop = stopwords.words('russian')
```

```python
pos_df = train_df[train_df['label'] == 1]
neg_df = train_df[train_df['label'] == 0]

train_df.loc[:, 'sentence'] = train_df['sentence'].apply(remove_stopwords_and_stem)
pos_df.loc[:, 'sentence'] = pos_df['sentence'].apply(remove_stopwords_and_stem)
neg_df.loc[:, 'sentence'] = neg_df['sentence'].apply(remove_stopwords_and_stem)
```

### Подготовка тренировочных данных.

Для начала подготовим наши тренировочные данные, разобьём их на обучающую/испытательную выборку.

```python
def my_preprocessor(text):
    """
    Делает предъобработку текста, удаляя лишние символы и прочий мусор от html.
    
    params:
    
    text [str]: - текст статьи.
    """
    text = re.sub('<[^>]*>', '', text)
    emotions = re.findall('(?::|;|=) (?:-)?(?:\)|\(|D|P)', text)
    text = (re.sub('[\W]+', ' ', text.lower()) + ' '.join(emotions).replace('-', ''))
    return text

def get_info_about_model(grid_search, scoring_type):
    """
    Получение информации о качествах метрик из сетки поиска.
    
    params:
    
    grid_search [GridSearchCV]: - обученный GridSearchCV.
    
    scoring_type [str]: - название метрики.
    """
    
    print(f"Лучшие параметры:\n{grid_search.best_params_}\n")
    print(f"[Train] {scoring_type}: {np.round(grid_search.best_score_, 2)}")
    
    if scoring_type == 'Precision':
        print(f"[Test] {scoring_type}: {np.round(precision_score(y_test, grid_search.best_estimator_.predict(X_test)), 3)}")
        print(f"[Test] Recall: {np.round(recall_score(y_test, grid_search.best_estimator_.predict(X_test)), 3)}")
    elif scoring_type == 'ROC_AUC':
        print(f"[Test] {scoring_type}: {np.round(roc_auc_score(y_test, grid_search.best_estimator_.predict_proba(X_test)[:, 1]), 3)}")
        print(f"[Test] Precision: {np.round(precision_score(y_test, grid_search.best_estimator_.predict(X_test)), 3)}")

def check_model_metrics(clf, clf_type):
    """
    Получение информации о качествах метрик из модельки.
    
    params:
    
    clf [Object]: - обученная модель.
    
    clf_type [str]: - название алгоритма МО на основе которой училась модель.
    """
    print(f"[Train] Precision: {np.round(precision_score(y_train, clf.predict(X_train)), 2)}")
    print(f"[Test] Precision: {np.round(precision_score(y_test, clf.predict(X_test)), 2)}\n")
    
    print(f"[Train] Recall: {np.round(recall_score(y_train, clf.predict(X_train)), 2)}")
    print(f"[Test] Recall: {np.round(recall_score(y_test, clf.predict(X_test), 2))}\n")

    if clf_type != 'SGDClassifier':
        print(f"[Train] ROC-AUC: {np.round(roc_auc_score(y_train, clf.predict_proba(X_train)[:, 1]), 2)}")
        print(f"[Test] ROC-AUC: {np.round(roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1]), 2)}\n")
    else:
        print(f"[Train] ROC-AUC: {np.round(roc_auc_score(clf.decision_function(y_train)), 2)}")
        print(f"[Test] ROC-AUC: {np.round(roc_auc_score(clf.decision_function(y_test)), 2)}\n")        

def check_predict(clf):
    """
    Проверка предсказаний классификатора на 10 случайныйх записях.
    
    params:
    
    clf [Object] - Обученный классификатор.
    """
    check_sample = pd.concat([data[data['label'] == 0].sample(n=5), data[data['label'] == 1].sample(n=5)])
    check_sample.reset_index(inplace=True)
    for i in range(check_sample.shape[0]):
        print(f"Текст статьи:\n{check_sample['sentence'][i]}\n")
        print(f"Настоящая метка класса: {check_sample['label'][i]}")
        print(f"Предсказанная метка класса: {clf.predict([check_sample['sentence'][i]])}\n\n")
        
data = pd.read_csv('Исходники/train_data.csv')
data = data.sample(frac=1)
data.reset_index(drop=True, inplace=True)
X, y = data['sentence'], data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

#### [ЛР] Матрица ошибок.

```python
def plot_confusion_matrix(clf):
    """
    Создание и визуализация матрицы ошибок
    
    params:
    clf [Object]: - Обученный классификатор.
    """
    y_pred = clf.predict(X_test)
    cf_matrix = confusion_matrix(y_test, y_pred)

    group_names = ['TN','FP','FN','TP']
    group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]

    labels = [f"{v1}\n{v2}\n" for v1, v2 in zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)

    plt.figure(figsize=(8, 6), dpi=80)

    sns.heatmap(cf_matrix, annot=labels, fmt='')
    plt.title("Матрица ошибок", fontsize=16)
    plt.show()
```
